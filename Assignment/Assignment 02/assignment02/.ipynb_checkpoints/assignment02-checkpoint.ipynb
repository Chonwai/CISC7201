# Question 2 (25%)
* Write a function of Python that requires an argument of a year number, e.g., 2020. 
* The function is to return the number of days of the argument. 
* Invalid input should be pointed out "Invalid input".
* The program should be executed continuously.

```python
flag = 0
loop = 0
def checkdays(year):
    if int(year) % 4 == 0:
        return True
    else:
        return False
    
def checkInput(year):
    try:
        type(int(year))
        isInt = 1
    except:
        isInt = 0
    return isInt

# implement your logic
while loop == 0:
    year = input()
    flag = checkInput(year);
    if flag == 1:
        isArgument = checkdays(year)
        if isArgument == True:
            print('366 days')
        elif isArgument == False:
            print('365 days')
    elif flag == 0:
        print('Invalid input!')
```

# Question 3 (25%)
* Given a grade table in a Dictionary structure. 
  * Write a function to return the highest score, the lowest score, average score.
  * Write a function to generate a dictionary that uses 'class' as key and 'name' as values, e.g., {'one':'Alice, Bob, ...'}
  * Write the newly generated dictionary into a file "data.json" in [JSON format](https://stackabuse.com/reading-and-writing-json-to-a-file-in-python/).

```python
import json

grades = [
    {'name': 'Alice', 'class': 'one', 'score': 85},
    {'name': 'Bob', 'class': 'two', 'score': 95},
    {'name': 'Cindy', 'class': 'one', 'score': 90},
    {'name': 'Daniel', 'class': 'two', 'score': 90},
    {'name': 'Elisa', 'class': 'two', 'score': 80},
    {'name': 'Frank', 'class': 'one', 'score': 95}
]

def stat(grades):
    scoreList = list(map(lambda x: x['score'], grades))
    highest = max(scoreList)
    lowest = min(scoreList)
    totalScore = 0
    for item in scoreList:
        totalScore += item
    average = totalScore / len(scoreList)
    print(highest, lowest, average)
#     return highest, lowest, average

def generate(grades):
#     two = list(map(lambda x: x['class'], grades))
    classes = {}
    classOne = ''
    classTwo = ''
    for item in grades:
        if item['class'] == 'one':
            classOne += item['name']
            classOne += ' '
        elif item['class'] == 'two':
            classTwo += item['name']
            classTwo += ' '
    classes['one'] = classOne
    classes['two'] = classTwo
    return classes

stat(grades)

# implement your logic here

# write file in JSON format
with open('data.json', 'w') as outfile:
#     implement your logic here
    classes = generate(grades)
    json.dump(classes, outfile)
#     e.g., json.dump(data, outfile)
```

```python
import csv
import itertools

with open('sf-salaries/Salaries.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    L=[]
    for row in csv_reader:
        if (row["BasePay"] == ''):
            row["BasePay"] = 0
        if (row["OvertimePay"] == ''):
            row["OvertimePay"] = 0
        if (row["OtherPay"] == ''):
            row["OtherPay"] = 0
        if (row["Benefits"] == ''):
            row["Benefits"] = 0
        L.append([row["EmployeeName"],row["JobTitle"],row["BasePay"],row["OvertimePay"],row["OtherPay"],row["Benefits"],row["Year"],row["Status"]])
#     print(L[23]) # this will print the 23rd record of the csv file

# implement your answer 1 here
highestPay = 0
highestPayEmployee = ''
for item in L:
    try:
        if item[1] == 'TRANSIT SUPERVISOR':
            totalPay = float(item[5]) + float(item[2]) + float(item[3]) + float(item[4])
            if highestPay < totalPay:
                highestPay = totalPay
                highestPayEmployee = item[0]
        else:
            continue;
    except:
        continue;
print(highestPayEmployee, highestPay)

# implement your answer 2 here
highestPay = 0
highestPayEmployee = ''
for item in L:
    try:
        if (item[1].find('Fire Dept') != -1):
            totalPay = float(item[2]) + float(item[3]) + float(item[4])
            if highestPay < totalPay:
                highestPay = totalPay
                highestPayEmployee = item[0]
        else:
            continue;
    except:
        continue;
print(highestPayEmployee, highestPay)

# implement your answer 3 here
ptList = []
for item in L:
    if item[7] == 'PT':
        ptList.append(item)

def topN(list, N = 5):
    finalList = []
#     print(list[7892])
    for i in range(0, N):
        max1 = 0
        position = 0
        element = {}
        for j in range(0, len(list)):
            totalPay = float(list[j][2]) + float(list[j][3]) + float(list[j][4])
#             print(j ,totalPay)
            if max1 < totalPay:
                max1 = totalPay
                position = j
                element["EmployeeName"] = list[j][0]
                element["TotalPay"] = max1
        finalList.append(element)
        del list[position]
    print(finalList)

topN(ptList, 5)    
                
    
```

# Bonus Question (+10%)
* Randomly generate a list of 10,000 random numbers in [0,100000)
    * e.g., L = [1, 10, 2, 3, 7, 8, 13, 21, 34, 66, 14, ...]
* Write a program that print out all PRIME numbers in L
* Check [Sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

```python
# basic idea
import random

PRIMES = []
orginList = []

for i in range(10000):
    orginList.append(random.randrange(100000 - 1))
    
orginList.sort()
# print(orginList)

while len(orginList) != 0:
    tempList = []
    for j in range(1, len(orginList)):
#         print(orginList[0])
        if (int(orginList[j]) % int(orginList[0]) == 0):
#             print(orginList[j])
#             PRIMES.append(orginList[j])
            tempList.append(orginList[j])
    PRIMES.append(orginList[0])
    tempList.append(orginList[0])
#     print(len(PRIMES))
#     print(len(tempList))
    for temp in tempList:
        orginList.remove(temp)
#     print(len(orginList))
        
print(PRIMES)
        
        
```
